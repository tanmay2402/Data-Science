Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation
of 5 using Python. Interpret the results.

Ans.

```python
import scipy.stats as stats
import math

# Given values
mean = 50
std_dev = 5
n = 30  # assuming a sample size of 30

# Z value for 95% confidence interval
z_value = stats.norm.ppf(0.975)

# Standard error
standard_error = std_dev / math.sqrt(n)

# Confidence interval calculation
margin_of_error = z_value * standard_error
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

# Print the result
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag
matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20%
brown. Use Python to perform the test with a significance level of 0.05.

Ans.

```python
import numpy as np
import scipy.stats as stats

# Observed frequencies of M&M colors (example data)
observed = np.array([45, 50, 40, 25, 30, 55])

# Total number of M&Ms in the sample
total = observed.sum()

# Expected frequencies based on the distribution (percentage * total)
expected_percentages = np.array([0.20, 0.20, 0.20, 0.10, 0.10, 0.20])
expected = expected_percentages * total

# Perform the chi-square goodness-of-fit test
chi2_stat, p_value = stats.chisquare(observed, expected)

# Print the results
print(f"Chi-Square Statistic: {chi2_stat:.2f}")
print(f"P-Value: {p_value:.4f}")

# Decision based on the significance level (alpha = 0.05)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The distribution of colors does not match the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The distribution of colors matches the expected distribution.")
```

Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals
smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the
population who smoke.

Ans

```python
import math
import scipy.stats as stats

# Given data
n = 500  # Sample size
x = 60   # Number of smokers

# Sample proportion (p-hat)
p_hat = x / n

# Z value for 95% confidence interval (two-tailed)
z_value = stats.norm.ppf(0.975)

# Standard error
standard_error = math.sqrt((p_hat * (1 - p_hat)) / n)

# Margin of error
margin_of_error = z_value * standard_error

# Confidence interval
lower_bound = p_hat - margin_of_error
upper_bound = p_hat + margin_of_error

# Print the results
print(f"Sample proportion (p-hat): {p_hat:.4f}")
print(f"95% Confidence Interval: ({lower_bound:.4f}, {upper_bound:.4f})")
```

Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the
area corresponding to a chi-square statistic of 15.

Ans

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define the degrees of freedom
df = 10

# Define the chi-square statistic threshold
threshold = 15

# Generate values for the x-axis (range of chi-square values)
x = np.linspace(0, 30, 1000)

# Get the chi-square distribution values
y = stats.chi2.pdf(x, df)

# Create the plot
plt.figure(figsize=(8, 6))

# Plot the chi-square distribution
plt.plot(x, y, label=f'Chi-Square Distribution (df={df})', color='blue')

# Shade the area corresponding to chi-square statistic of 15 and beyond
plt.fill_between(x, y, where=(x >= threshold), color='skyblue', alpha=0.5, label=f'Area  {threshold}')

# Add labels and title
plt.title('Chi-Square Distribution with 10 Degrees of Freedom')
plt.xlabel('Chi-Square Value')
plt.ylabel('Density')

# Add a vertical line at the threshold
plt.axvline(x=threshold, color='red', linestyle='--', label=f'Threshold = {threshold}')

# Add a legend
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

Q7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520
preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who
prefer Coke.

Ans.

```python
import math
import scipy.stats as stats

# Given data
n = 1000        # Sample size
x = 520         # Number of people who prefer Coke

# Sample proportion (p-hat)
p_hat = x / n

# Z value for 99% confidence interval (two-tailed)
z_value = stats.norm.ppf(0.995)  # For 99% CI, alpha = 0.01, so alpha/2 = 0.005

# Standard error
standard_error = math.sqrt((p_hat * (1 - p_hat)) / n)

# Margin of error
margin_of_error = z_value * standard_error

# Confidence interval
lower_bound = p_hat - margin_of_error
upper_bound = p_hat + margin_of_error

# Print the results
print(f"Sample proportion (p-hat): {p_hat:.4f}")
print(f"99% Confidence Interval: ({lower_bound:.4f}, {upper_bound:.4f})")
```

Q11. A random sample of 30 people was selected from a population with an unknown mean and standard
deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10.
Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a
significance level of 0.05.

Ans

To conduct a hypothesis test to determine if the population mean is significantly different from 70, we will use a **t-test** since the population standard deviation is unknown, and we are dealing with a sample. Here are the steps for conducting the hypothesis test:

### Step 1: State the hypotheses
- **Null hypothesis (H):** The population mean is 70, i.e., \( \mu = 70 \).
- **Alternative hypothesis (H):** The population mean is different from 70, i.e., \( \mu \neq 70 \).

### Step 2: Set the significance level
The significance level () is given as 0.05.

### Step 3: Gather the sample information
- Sample size (n) = 30
- Sample mean ( \( \bar{x} \) ) = 72
- Sample standard deviation (s) = 10
- Population mean under the null hypothesis ( \( \mu_0 \) ) = 70

### Step 4: Calculate the t-test statistic
The formula for the t-statistic is:

\[
t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}}
\]

Where:
- \( \bar{x} \) is the sample mean,
- \( \mu_0 \) is the population mean under the null hypothesis,
- \( s \) is the sample standard deviation,
- \( n \) is the sample size.

Plugging in the values:

\[
t = \frac{72 - 70}{10 / \sqrt{30}}
\]

\[
t = \frac{2}{10 / \sqrt{30}} = \frac{2}{1.8257} \approx 1.095
\]

So, the calculated t-statistic is approximately **1.095**.

### Step 5: Determine the degrees of freedom
The degrees of freedom (df) for a t-test are calculated as:

\[
df = n - 1
\]

For this sample:

\[
df = 30 - 1 = 29
\]

### Step 6: Find the critical t-value
Since this is a **two-tailed test**, we will look for the critical t-value corresponding to \( \alpha = 0.05 \) with \( df = 29 \). You can find this value using a t-distribution table or a calculator. 

For \( \alpha = 0.05 \) (split between the two tails), the critical t-value for 29 degrees of freedom is approximately **2.045**.

### Step 7: Make a decision
Now, compare the absolute value of the calculated t-statistic to the critical t-value:
- If \( |t_{\text{calculated}}| > t_{\text{critical}} \), reject the null hypothesis.
- If \( |t_{\text{calculated}}| \leq t_{\text{critical}} \), fail to reject the null hypothesis.

From our calculations:
- \( |t_{\text{calculated}}| = 1.095 \)
- \( t_{\text{critical}} = 2.045 \)

Since \( 1.095 < 2.045 \), we **fail to reject the null hypothesis**.

### Step 8: Conclusion
At the 0.05 significance level, there is insufficient evidence to conclude that the population mean is significantly different from 70. The sample mean of 72 is not significantly different from 70 based on the data provided.
